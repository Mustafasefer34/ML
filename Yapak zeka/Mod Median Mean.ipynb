Mean Median ve Mod örnek uygulamalar

```python
import numpy as np
```

```python
#Aylık gelir 5000 standart sapma 2000 ve 1000 kişi
gelirler = np.random.normal(5000,2000,1000)
gelirler
```

```python
np.mean(gelirler)
```

```python
import matplotlib.pyplot as plt
plt.hist(gelirler,100)
plt.show()
```

```python
np.median(gelirler)
```

```python
#Bill gates tesadüfen veri setimizin içine girerse
gelirler=np.append(gelirler,[100000000])
np.mean(gelirler)
```

```python
np.median(gelirler)
```

```python
np.median(gelirler) #median daha doğru sonucu verebilir
```

```python
yaslar=np.random.randint(7,18,size=300)
from scipy import stats
stats.mode(yaslar)
#15 yaş 37 kere tekrar etmiş
```

```python
np.mean(yaslar)
```

```python
np.median(yaslar)
```
