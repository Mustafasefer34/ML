```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.show()
```

```python
plt.title("lk Grafiimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("lk Grafiimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,label='x^2',color='green')
plt.xticks([1,2,3,4,5])
plt.yticks([1,4,9,16,25])
plt.title("lk Grafiimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label='x^2',color='blue',linewidth=2,linestyle='--',marker='.')
plt.xticks([1,2,3,4,5])
plt.yticks([1,4,9,16,25])
plt.title("lk Grafiimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label='x^2',color='blue',linewidth=2,linestyle='--',marker='.')
plt.xticks([1,2,3,4,5])
plt.yticks([1,4,9,16,25])
plt.title("lk Grafiimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")

x2=np.arange(0,5,0.5)
plt.plot(x2,x2*2,color='red',linewidth=2,marker='.',label='2*x')
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label='x^2',color='blue',linewidth=2,linestyle='--',marker='.')
plt.xticks([1,2,3,4,5])
plt.yticks([1,4,9,16,25])
plt.title("lk Grafiimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")

x2=np.arange(0,5,0.5)
plt.plot(x2,x2*2,color='red',linewidth=2,marker='.',label='2*x')
plt.legend()
plt.savefig('ilkfgrafiim.png',dpi=300)
plt.show()
```

```python
gas=pd.read_csv('petrolfiyatlari.csv')
gas
```

```python
plt.title('Petrol Fiyatlar')
plt.plot(gas['Year'],gas['USA'],'b-',label='USA')
plt.xlabel('Yl')
plt.ylabel('Dolar')
plt.legend()
plt.show()
```

```python
plt.title('Petrol Fiyatlar')
plt.plot(gas.Year,gas.USA,'b-',label='USA')
plt.plot(gas.Year,gas['Canada'],'r-',label='Kanada')
plt.plot(gas.Year,gas['South Korea'],'g.-',label='Gney Kore')
plt.plot(gas.Year,gas['France'],'y.-',label='Fransa')
plt.xlabel('Yl')
plt.ylabel('Dolar')
plt.legend()
plt.show()
```
